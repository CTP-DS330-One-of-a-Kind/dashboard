```python
import streamlit as st
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Data Cleaning
df = pd.read_csv("data/MW.csv", encoding='cp1252')
df_state = pd.read_csv("data/state.csv", encoding="unicode_escape")
df['State'] = df['State'].str.lower()
df.columns = df.columns.str.lower()
merged_df = pd.merge(df,df_state[['state','year','governor']])
```

```python
df.head()
```

```python
states = gpd.read_file('shape/cb_2018_us_state_5m.shp')
```

```python
states.head()
```

```python
filtered_df = merged_df[merged_df['year'] == 1999]
```

```python
filtered_df.head()
```

```python
filtered_df_unmerged = df[df['year'] == 1999]
```

```python
filtered_df_unmerged.head()
```

```python
states['wage'] = states['NAME'].str.lower().map(filtered_df_unmerged.set_index('state')['effective.minimum.wage.2020.dollars'])
```

```python
plt.ioff()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))  # Adjust the figsize as needed
#ax.set_axis_off()
ax.set_xlim(-126, -66) 
ax.set_ylim(24, 50)
```

```python
states.plot(ax=ax, column='wage',
            linewidth=0.5, edgecolor='0.8', 
            cmap='viridis', legend=True)
plt.show()
```
